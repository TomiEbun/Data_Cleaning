####  Virtual assessment for data cleaning:

 4) Format the release year date to be consistent uisng datetime function
 
 5) Split the genere and select the first genre for each movie

 8) Remove non digit values from the income
 
 9) Remove all leading and trailing spaces 

```python
# To convert columns name to lowercase
md.columns=md.columns.str.lower()
```

```python
md.head(5)
```

```python
## To remove trailing spaces from the column names 
md.columns.str.replace(' ','_')
```

```python
#To rename Columns 
md.rename(columns={'imbd title id' :'title_id','original titlê' :'original_title','content rating' :'content_rating','genrë¨' :'genre','release year' :'release_year'},inplace=True)
```

```python
### To drop columns
md.drop(md.columns[[0, 9,5,11]], axis=1, inplace=True)
```

```python
##To split and reassigning back to the column
md['title_id']=md['title_id'].str.split('tt',expand=True)[1]
```

```python
md.head()
```

```python
md.release_year
```

```python
##To change the date format
md.release_year = pd.to_datetime(md.release_year, errors='coerce')
```

```python
md.info()
```

```python
md.head(5)
```

```python
### To split the genre column
md['genre'].str.split(',',expand=True)
```

```python
#Assign the splited genre to its main column
md['genre']=md['genre'].str.split(',',expand=True)[0]
```

```python
### To replace cells in a country column 
md['country']=md['country'].replace(['US','New Zeland','New Zesland','West Germany'],
                                    ['USA','New Zealand','New Zealand','Germany'])
```

```python
### To replace content_rating cells
md['content_rating'] = md['content_rating'].replace(['Approved', 'Unrated','G','PG','R'], ['Rated','Not Rated','PG-13','PG-13','Rated'])
```

```python
##To check the income column
md.income
```

```python
#To replace the $ and trailing spaces in income column
md.income = md.income.str.lstrip('$  ')
```

```python
md.income = md.income.str.replace(',', '')
```

```python
md.income = md.income.str.replace('o', '0')
```

```python
#To check the position of Nan in the cell
unique_values = md['income'].unique()
print(unique_values)
```

```python
## To replace the row with Nan
md.loc[13,'income']=0
```

```python
md.income.dtypes
```

```python
#To change income from object datatype to float
md.income = md.income.astype(float)
```

```python
##To change Income from Float datatype to Integer
md.income = md.income.astype(int)
```

```python
md.info()
```

```python
### To add a dollar sign $ and comma to the integer values 
md.income = md.income.apply(lambda x: '${:,.0f}'.format(x))
```
